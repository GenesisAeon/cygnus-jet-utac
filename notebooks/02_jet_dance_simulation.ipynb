# Notebook 02: 18-Year Jet Dance Simulation

**GenesisAeon Package 17** — full 18-year UTAC simulation matching the Prabu et al. 2026 observation baseline.

This notebook reproduces the 'dancing' jet deflection detected over 18 years of VLBI monitoring (2006–2024).

```python
import numpy as np
import matplotlib.pyplot as plt
from cygnus_jet_utac import CygnusJetUTAC
from cygnus_jet_utac.constants import LIGHT_YEAR

np.random.seed(42)

# Full 18-year simulation with 6-hour steps
system = CygnusJetUTAC(dt=21600.0, seed=42)
print('Running 18-year simulation...')
results = system.run_cycle(duration_years=18.0)
print(f'Done. {results["n_steps"]:,} steps, {results["n_dance_events"]} dance events detected.')
```

## Dance Events Timeline

```python
system.plot_phase_events(save_path='../docs/dance_events_18yr.png')

events = system.get_phase_events()
print(f'\n=== Dance Events ({len(events)} total) ===')
for ev in events[:10]:
    print(f'  t = {ev["t_years"]:.2f} yr  |  deflection = {ev["deflection_angle_deg"]:.1f}°  |  Γ = {ev["crep_gamma"]:.4f}')
if len(events) > 10:
    print(f'  ... and {len(events)-10} more events')
```

## Falsifiable Prediction

The UTAC model predicts the next jet dance event based on the last detected phase transition.

```python
from cygnus_jet_utac.constants import CYG_ORBITAL_PERIOD

if events:
    last_event = events[-1]
    t_last = last_event['t_years']
    T_orb_yr = CYG_ORBITAL_PERIOD / (365.25 * 86400)
    
    # Prediction: next event within 1.5–2 orbital periods
    t_pred_min = t_last + 1.5 * T_orb_yr
    t_pred_max = t_last + 3.0 * T_orb_yr
    
    obs_year_min = 2006 + t_pred_min
    obs_year_max = 2006 + t_pred_max
    
    print('=== FALSIFIABLE PREDICTION ===')
    print(f'  Last dance event:  t = {t_last:.2f} yr (year {2006+t_last:.1f})')
    print(f'  Next event window: {obs_year_min:.1f} – {obs_year_max:.1f}')
    print(f'  (1.5–3 orbital periods = {1.5*T_orb_yr*365.25:.0f}–{3.0*T_orb_yr*365.25:.0f} days)')
    print()
    print('  Test with future VLBI monitoring of Cygnus X-1.')
else:
    print('No dance events detected — try a longer simulation.')
```
